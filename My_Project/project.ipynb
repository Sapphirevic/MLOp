```python
plt.hist(new_2['salary_in_usd'], bins=20)
plt.xlabel('Salary_in_usd')
plt.ylabel('Frequency')
plt.title('Distribution of Work Year')
plt.show()
```

```python
sns.boxplot(x='salary_in_usd', y='employment_type', data=new_2)
plt.xlabel('Salary Currency')
plt.ylabel('employment_type')
plt.title('Boxplot of Salary by Employment')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='job_category', data=new_2, order=new_2['job_category'].value_counts().index)
plt.xlabel('count')
plt.ylabel('Job Category')
plt.title('Distribution of Job Categories')
plt.show()
```

```python
sns.countplot(x='experience_level', data=new_2)
plt.xlabel('Experience Level')
plt.ylabel('Count')
plt.title('Count of Experience Level')
plt.show()
```

```python
plt.figure(figsize=(20, 8))
plt.scatter(new_2['company_location'], new_2['salary'])
plt.xlabel('Company Location')
plt.ylabel('Salary')
plt.title('Company Location vs Salary')
plt.xticks(rotation=45)
plt.show()
```
